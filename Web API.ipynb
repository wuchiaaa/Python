```python
'''
透過網站上公開的API資料點，用python存取這些API
'''
```

```python
# EXAMPLE_01.GitHub上查詢個人主要訊息
# https://api.github.com/users/用户名
```

```python
import requests
import pandas as pd
url_01 = 'https://api.github.com/users/wuchiaaa'
resp = requests.get(url_01)
resp
```

```python
data_01 = resp.json()
data_01
```

```python
info = pd.DataFrame(data_01, index = [0], columns = ['login', 'name', 'public_repos'])
info
```

```python
# EXAMPLE_02.GitHub上查詢(帳號來自pandas-dev)pandas最新的5個issue
# https://api.github.com/repos/用户名/仓库名/issues
url_02 = 'https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url_02)
resp
```

```python
data_02 = resp.json()
data_02
```

```python
issues = pd.DataFrame(data_02, columns=['number', 'title', 'updated_at', 'state'])
issues.head()
```
